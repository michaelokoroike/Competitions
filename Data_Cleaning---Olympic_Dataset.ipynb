# Columns - Change Them
   * Looks like the column names we could use are actually in the first row, when looking at it; make these the column names
   
   * Had to devise a way since I could not just use locals
   
   * First, we want the first row to be the column names for the dataset
   
   * Because it has been difficult to simply access this row, we will use the tolist method to access it.

```python
x = 0 #will end up representing rows being iterated through

while x < 1: #before we get to the 2nd row
    rows = []
    for row in df.values.tolist(): #this is a list of every row
        #print(rows)
        rows.append(row) #append first row of the dataframe to the empty list
        x += 1
        if x == 1:
            break

df.columns = rows

df.head()
```

# Columns Names - What do they represent?
* What I believe each column represents:

    * Summer = Amount of summer olympic games the country has competed in
    
    * Winter = Amount of winter olympic games the country has competed in
    
    * Games = Total number of summer and winter olympic games the country has competed in
    
    * 01!, 02!, 03! = I am guessing (no data out there on this dataset) that these mean gold, silver, and bronze (totals)
    
    * Total = Total medals in summer and winter games
    
    * Combined_Total = Combined total amount of medals the country has won in the olympics

* For the purposes of data cleaning, let us simply make this all clearer to understand

* Let's rename them
    
    * Take parentheses and commas away
    
    * The ones that do not need to be determined by position in column names:
        
        * Summer = Summer_Olympics_Competed_In
        
        * Winter = Winter_Olympics_Competed_In
        
        * Games = Total_Olympics_Competed_In
        
        * Combined_Total = Total_Medals_Earned
    
    * The ones that will depend on value
        
        * 01!, 02!, 03! = Some will be Golds_Summer, some will be Golds_Winter, some will be Golds_Total
        
        * Total = Total_Medals_Earned_Summer, Total_Metals_Earned_Winter
        

```python
df = df.rename(columns={"Summer":"Summer_Olympics_Competed_In", "Winter":"Winter_Olympics_Competed_In",
                        "Games":"Total_Olympics_Competed_In", "(Combined_Total,)":"Total_Medals_Earned"})
```
